```python
import h5py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf
```

# Carga de datasets

```python
def cargar_datasets_desde_h5(archivo_h5):
    datasets = {}
    with h5py.File(archivo_h5, "r") as f:
        for grupo in f:
            datasets[grupo] = {}
            for key in f[grupo]:
                datasets[grupo][key] = f[grupo][key][:]
    return datasets

datasets_train = cargar_datasets_desde_h5("datasets_train.h5")
```

# Dataset A

Analisis EDA

```python
X_train = datasets_train['A']['X_train']
Y_train = datasets_train['A']['y_train']
X_test = datasets_train['A']['X_test']

#E1.1
print("Régimen: \nn:"+str(X_train.shape[0])+ "\np:"+str(X_train.shape[1]))
```

Respuesta 1.1:

```python
#E1.2

# Matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(pd.DataFrame(X_train).corr(), cmap='coolwarm', center=0, annot=False)
plt.title("Matriz de Correlación")
plt.show()

# Gráfico de Autocorrelación (ACF)
plt.figure(figsize=(10, 4))
plot_acf(pd.DataFrame(X_train).iloc[:, 0], lags=40, ax=plt.gca())
plt.title("Autocorrelación de la Feature 0")
plt.show()
```

```python
#E1.3

# Boxplot para identificar outliers
sns.boxplot(data=pd.DataFrame(X_train).iloc[:, :5]) 
plt.title("Detección de Outliers")
plt.show()
```
