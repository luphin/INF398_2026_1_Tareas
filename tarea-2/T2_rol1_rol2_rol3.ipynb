# Carga de datasets

```python
def cargar_datasets_desde_h5(archivo_h5):
    datasets = {}
    with h5py.File(archivo_h5, "r") as f:
        for grupo in f:
            datasets[grupo] = {}
            for key in f[grupo]:
                datasets[grupo][key] = f[grupo][key][:]
    return datasets

datasets_train = cargar_datasets_desde_h5("datasets_train.h5")
print("Estructura de datasets_train.h5:")
for letter in "ABCDE":
    keys = list(datasets_train[letter].keys())
    print(f"  {letter}: keys={keys}")
    for k in keys:
        arr = datasets_train[letter][k]
        print(f"     {k}: shape={arr.shape} dtype={arr.dtype}")
```
